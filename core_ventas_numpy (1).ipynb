```python
import numpy as np
from numpy import genfromtxt
from google.colab import drive

def cargar_datos(ruta_archivo):
    # Carga los datos del archivo CSV utilizando NumPy
    datos = np.genfromtxt(ruta_archivo, delimiter=",",dtype= [('Transaction ID',int ),('Date',str), ('Customer ID', 'U10'),
    ('Gender', 'U10'),('Age', int),('Product Category', 'U16'), ('Quantity', int),('Price', float),('Total Amount', int)])
    datos = np.array(datos)
    return datos

def calcular_ventas_totatles(datos):
    # Calcula el total de ventas
    ventas_totales = np.sum(datos['Total_Amount'])
    return ventas_totales

def calcular_ventas_por_genero(datos):
    # Calcula el total de ventas por genero
    columna_genero = datos['Gender']
    columna_ventas = datos['Total_Amount']
    mascara = columna_genero == 'Male'
    ventas_por_genero = np.sum(columna_ventas[mascara])
    return ventas_por_genero

def calcular_ventas_por_producto(datos):
    # Calcula el total de ventas por producto
    columna_producto = datos['Product_Category']
    columna_ventas = datos['Total_Amount']
    producto1 = columna_producto == 'Electronics'
    ventas_producto1 = np.sum(columna_ventas[producto1])
    promedio_producto1 = np.mean(columna_ventas[producto1])
    producto2 = columna_producto == 'Clothing'
    producto3 = columna_producto == 'Beauty'
    ventas_producto2 = np.sum(columna_ventas[producto2])
    promedio_producto2 = np.mean(columna_ventas[producto2])
    ventas_producto3 = np.sum(columna_ventas[producto3])
    promedio_producto3 = np.mean(columna_ventas[producto3])
    print("ventas de Electronics:",ventas_producto1)
    print("promedio de Electronics:",promedio_producto1)
    print("ventas de Clothing:",ventas_producto2)
    print("promedio de Clothing:",promedio_producto2)
    print("ventas de Beauty:",ventas_producto3)
    print("promedio de Beauty:",promedio_producto3)

    if ventas_producto1 > ventas_producto2 and ventas_producto1 > ventas_producto3:
        print("El producto mas vendido es Electronics")

    elif ventas_producto2 > ventas_producto1 and ventas_producto2 > ventas_producto3:
        print("El producto mas vendido es Clothing")

    else:
        print("El producto mas vendido es Beauty")

    return ventas_producto1, ventas_producto2, ventas_producto3


if __name__ == "__main__":

    ruta_archivo = '/content/drive/My Drive/Colab Notebooks/retail_sales_dataset.csv'
    datos = cargar_datos(ruta_archivo)
    print(datos)
    ventas_totales = calcular_ventas_totatles(datos)
    print()
    print("ventas totales:", ventas_totales)
    Ventas_por_genero = calcular_ventas_por_genero(datos)
    print("ventas Hombres:", Ventas_por_genero)
    ventas_mujeres = ventas_totales - Ventas_por_genero
    print("ventas Mujeres:", ventas_mujeres)
    ventas_por_producto = calcular_ventas_por_producto(datos)

```
